```
df = my_spark.read.format("mongo").option("spark.mongodb.input.uri", "mongodb+srv://admin:12345678Pc@cluster0.k5xvg.mongodb.net/2020MSBD5003GP11.brandcounts?retryWrites=true&w=majority").load()
```
